## 1. Check LSTM vs LSTM+Attention

```python
import pandas as pd

summary = dict()
for dataset in ['esnli', 'yelphat50', 'hatexplain']:
    summary[dataset] = pd.read_parquet(f'/Users/dunguyen/Projects/IGRIDA/historic/2022-12-13/summary/overfitting_check/lstm_vanilla/{dataset}/summary.parquet')
```

```python
df_summary = pd.concat(list(summary.values()))
df_summary = df_summary.groupby(['data', 'n_context'])[['y:fscore']].agg(['mean', 'sem'])

df_summary['y:fscore','ci95_hi'] = df_summary['y:fscore', 'mean'] + 1.96* df_summary['y:fscore','sem']
df_summary['y:fscore','ci95_lo'] = df_summary['y:fscore', 'mean'] + 1.96* df_summary['y:fscore','sem']
df_summary['y:fscore','value'] = df_summary['y:fscore', 'mean'].round(3).map(str) + '±' + (1.96* df_summary['y:fscore','sem']).round(3).map(str)
df_summary
```

```python

summary = dict()
for dataset in ['esnli', 'yelphat50', 'hatexplain']:
    summary[dataset] = pd.read_parquet(f'/Users/dunguyen/Projects/IGRIDA/historic/2022-12-13/summary/overfitting_check/embedding_dimension_variation/{dataset}/summary.parquet')
    
df_summary = pd.concat(list(summary.values()))

metric = 'y:fscore'
df_summary = df_summary.groupby(['data', 'n_context'])[[metric]].agg(['mean', 'sem'])

df_summary[metric,'ci95_hi'] = df_summary[metric, 'mean'] + 1.96* df_summary[metric,'sem']
df_summary[metric,'ci95_lo'] = df_summary[metric, 'mean'] + 1.96* df_summary[metric,'sem']
df_summary[metric,'value'] = df_summary[metric, 'mean'].round(3).map(str) + '±' + (1.96* df_summary[metric,'sem']).round(3).map(str)
display(HTML(f'<h3>{metric}</h3>'))
df_summary
```
